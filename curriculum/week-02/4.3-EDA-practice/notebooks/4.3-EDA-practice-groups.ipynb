```python
# First we need to fetch some data using Python requests from API
# Read more about Python requests:
# http://docs.python-requests.org/en/master/user/quickstart/

import requests

response = requests.get("http://api.dronestre.am/data")
json_data = response.json()
drone_df = pd.DataFrame(json_data['strike'])
```

```python
alt_drone = drone_df
```

```python
alt_drone.index = pd.to_datetime(alt_drone['date'])
```

```python
alt_drone['lat'] = alt_drone['lat'].map(lambda value: float(value) if len(value) > 0 else np.nan).astype(float)
alt_drone['lon'] = alt_drone['lon'].map(lambda value: float(value) if len(value) > 0 else np.nan).astype(float)
```

```python
# alt_drone[['deaths_min', 'deaths_max']].info()
alt_drone['deaths_max'] = alt_drone['deaths_max'].map(lambda value: int(value) if value.replace("?", "") != "" else np.nan).astype(float)
```

```python
import folium

# data cleaning
alt_drone['lat'] = alt_drone['lat'].map(lambda value: float(value) if len(value) > 0 else np.nan).astype(float)
alt_drone['lon'] = alt_drone['lon'].map(lambda value: float(value) if len(value) > 0 else np.nan).astype(float)
alt_drone['deaths_max'] = alt_drone['deaths_max'].map(lambda value: int(value) if value.replace("?", "") != "" else np.nan).astype(float)

# Convert date/time to index for plotting
alt_drone.index = pd.to_datetime(alt_drone['date'])

drone_map = folium.Map(location=[34.81, 70.48], tiles='Stamen Terrain',
                   zoom_start=4)

def add_marker(row):
    
    title = "%s, %s\n%s" % (
        row['location'], 
        row['town'], 
        row['bij_summary_short']
    )
    
    if row['deaths_max'] > 100:
        color = "red"
    elif row['deaths_max'] > 50:
        color = "yellow"
    elif row['deaths_max'] > 15:
        color = "green"
    else:
        color = "grey"
    
    folium.Marker(
        location =  [row['lat'], row['lon']], 
        popup    =  title,
        icon     =  folium.Icon(icon = 'cloud', color=color)
    ).add_to(drone_map)

alt_drone.head(200).apply(add_marker, axis=1)


drone_map

#alt_drone.resample("M")['deaths_max'].sum().plot()
```

```python
alt_drone.groupby('target').size().sort_values(ascending=False).head(10)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation

vect = CountVectorizer(stop_words="english")
X = vect.fit_transform(alt_drone['bij_summary_short'])

# model params
n_samples = 2000
n_features = 1000
n_topics = 4
n_top_words = 5

lda = LatentDirichletAllocation(
    n_topics=n_topics, 
    max_iter=5,
    learning_method='online',
    learning_offset=50.,
    random_state=0
)

model = lda.fit(X)

def print_topic_words(model, features):

    for topic_id, topic in enumerate(model.components_):
        print "\n\nTopic %d\n====================" % topic_id
        print(sorted(zip(features, topic), key=lambda val: val[1], reverse = True)[:10])

    
print_topic_words(model, vect.get_feature_names())

# def print_top_words(model, feature_names, n_top_words):
#     for topic_idx, topic in enumerate(model.components_):
#         print("Topic #%d:" % topic_idx)
#         print(" ".join([feature_names[i]
#                         for i in topic.argsort()[:-n_top_words - 1:-1]]))
#     print()
    
# print_top_words(model, vect.get_feature_names(), n_features)
```
