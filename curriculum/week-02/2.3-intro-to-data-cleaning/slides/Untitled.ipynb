```python
import numpy as np
import pandas as pd

rock_songs_file = '/Users/tyronewilliams/Desktop/DSI-SF-4/datasets/rock_songs/rock.csv'
rock_songs=pd.read_csv(rock_songs_file)
rock_songs.head(5)


column_names ={"Song Clean": "Song",
               "ARTIST CLEAN": "Artist",
               "COMBINED": "Song/Artist",
               "First?": "First",
               "Year?": "Year" }
rock_songs.rename(columns= column_names, inplace=True)

rock_songs.describe()

rock_songs["Year"].value_counts()







```

```python
#**1. What are the top 20 most popular songs by plays?**
rock_songs.sort(columns=["PlayCount"], ascending = False).head(20)
```

```python
#**2. Which years have the most plays?**
rock_songs["Release Year"].value_counts().head(5)
```

```python
#**3. Which records don't have matching "Play Count" cooresponding to "F*G"?**
def checker(inp):
    if (inp["Year"] * inp["PlayCount"])==inp["F*G"]:
        return True
    
rock_songs.apply(checker)
```
