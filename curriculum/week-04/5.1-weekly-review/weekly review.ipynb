```python
# df.groupby('Year').agg({'TenYD':np.min, 'Weight':np.max})
# [['Name', 'Position', 'PickTotal']]

has_weight = df.Weight > 0
has_yd = df.TenYD > 0


def max_weight(subdf):
    subdf = subdf.sort_values('Weight', ascending=False)
    return subdf.iloc[0,:]

def min_speed(subdf):
    subdf = subdf.sort_values('TenYD', ascending=True)
    return subdf.iloc[0,:]

weights = df[df.Weight > 0].groupby('Year').apply(max_weight).reset_index(drop=True)
speeds = df[df.TenYD > 0].groupby('Year').apply(min_speed).reset_index(drop=True)

combined = pd.concat([weights, speeds], axis=0)[['Year','Position','Name','TenYD','Weight','PickTotal']]
```
