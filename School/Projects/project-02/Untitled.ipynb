```python
def level(x):
    dict1={"A":"Boat Deck", "F": "Upper Deck",
           "B": "Promenade Deck", "G":"Middle Deck", "C":"Bridge Deck","D":"Shelter Deck", "E":"Saloon Deck"}
    try:
        return dict1[x[0]]
    except:
        return np.nan

titanic["Levels"] = titanic["Cabin"].map(level)

```
